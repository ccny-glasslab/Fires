```python
import numpy as np
import matplotlib.pyplot as plt


N = 4
menMeans = (1.654863, 1.496526,1.08069, 1.342072)
menStd = (0.108271, 0.083871, 0.119302, 0.097117)

fig, ax = plt.subplots()

ind = np.arange(N)    # the x locations for the groups
width = 0.35         # the width of the bars
p1 = ax.bar(ind, menMeans, width, bottom=0, color = 'teal', yerr=menStd)


womenMeans = (2.084102, 1.856303, 1.42902, 1.657210)
womenStd = (0.217312, 0.172224, 0.222992, 0.177363)
p2 = ax.bar(ind + width, womenMeans, width, bottom=0, color = 'olive', yerr=womenStd)

ax.set_title('Error Values for Super Resolution')
ax.set_xticks(ind + width / 2)
ax.set_xticklabels(('Baseline', 'Laplace', 'Unsharp Mask', 'ESRGAN'))

ax.legend((p1[0], p2[0]), ('Band 7', 'Band 14'))
ax.autoscale_view()

plt.show()
```
