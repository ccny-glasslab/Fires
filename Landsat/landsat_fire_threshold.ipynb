```python
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
bt039 = []
bt103 = []
landsat_path = '../landsat_files/landsat_files/'

for landsat_file in os.listdir(landsat_path):
    if landsat_file[63:-4] == '0.02deg':
        bt039 = bt039 + list(np.load(landsat_path + landsat_file).flatten())
        print(landsat_file)
    if landsat_file[64:-4] == '0.02deg':
        bt103 = bt103 + list(np.load(landsat_path + landsat_file).flatten())
```

```python
img039 = np.load(landsat_path + 'LC08_L2SP_044032_20210828_20210901_02_T1_goesLS_BT3.9_gridBcent0.02deg.npy')
img103 = np.load(landsat_path + 'LC08_L2SP_044032_20210828_20210901_02_T1_goesLS_BT10.3_gridBcent0.02deg.npy')
diff = img039 - img103
diff_masked = diff.copy()
diff_masked[diff < 1] = 0

f, ax = plt.subplots(1, 3, figsize=(20, 5))
ax[0].imshow(diff)
ax[1].imshow(diff_masked)
ax[2].scatter(img039, img103)
```
