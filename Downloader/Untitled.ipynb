```python
import xarray
import numpy as np
import matplotlib.pyplot as plt
import metpy
import cartopy.crs as ccrs
```

```python
data = xarray.open_dataset('../../GOES_Files/test_files/OR_ABI-L2-MCMIPC-M3_G17_s20183171802219_e20183171804592_c20183171805082.nc')
```

# RGB Image

```python
R = data['CMI_C02'].data
G = data['CMI_C03'].data
B = data['CMI_C01'].data

R = np.clip(R, 0, 1)
G = np.clip(G, 0, 1)
B = np.clip(B, 0, 1)

gamma = 2.2
R = np.power(R, 1/gamma)
G = np.power(G, 1/gamma)
B = np.power(B, 1/gamma)

G_true = 0.45 * R + 0.1 * G + 0.45 * B
G_true = np.maximum(G_true, 0)
G_true = np.minimum(G_true, 1)
    
RGB = np.dstack([R, G_true, B])
```

```python
fig, ax = plt.subplots()

ax.imshow(RGB)
ax.set_title('GOES-17 RGB 11/13/2018', fontweight='semibold', loc='left', fontsize=12);
ax.axis('off')
```

```python
dat = data.metpy.parse_cf('CMI_C02')
geos = dat.metpy.cartopy_crs

x = dat.x
y = dat.y

pc = ccrs.PlateCarree()
```

```python
fig = plt.figure(figsize=(10, 10))

ax = fig.add_subplot(projection=pc)

ax.set_extent([-124.50, -114.25, 32.50, 42.00], crs=pc)

ax.imshow(RGB, origin='upper',
           extent=(x.min(), x.max(), y.min(), y.max()),
           transform=geos,
           interpolation='none')
ax.add_feature(ccrs.cartopy.feature.STATES)
ax.set_title('CA 11/13/2018', loc='left', fontweight='bold')
```
