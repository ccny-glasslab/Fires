```python
import xarray as xr
import metpy
import cartopy.crs as ccrs
from pyresample import geometry, grid
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from scipy import stats
import os
```

```python
goes_params = geos.proj4_params

pc = ccrs.PlateCarree()
pc_extents = [-124.25, -114.25, 32.25, 42.25]
pc_params = pc.proj4_params
target_extents = [pc_extents[0],
                    pc_extents[2],
                    pc_extents[1],
                    pc_extents[3]]
target_rows = 500
target_cols = 500
shape = [target_rows, target_cols]

def create_npy(array, file):
    x = np.asarray(array)
    source_area = geometry.AreaDefinition.from_cf(file)
    target_area = geometry.AreaDefinition.from_extent('CA', pc_params, shape, target_extents)
    return(grid.get_resampled_image(target_area, source_area, x))
```

# All the Bands

```python
data = xr.open_dataset('../GOES_Files/test_files/OR_ABI-L2-MCMIPC-M3_G17_s20183171802219_e20183171804592_c20183171805082.nc')
for n in range (1,17):
    if n < 10:
        band = '0' + str(n)
    else:
        band = str(n)
        
    dat = data.metpy.parse_cf('CMI_C' + band)
    geos = dat.metpy.cartopy_crs
    rad = dat.data

    np.save('../GOES_Files/test_files/multiband_npys/' + band, create_npy(rad, data))
    data.close()
```

# 7 and 14 Brightness Temperatures

```python
myFile = xr.open_dataset('../GOES_Files/test_files/OR_ABI-L1b-RadC-M3C07_G17_s20183171802219_e20183171805003_c20183171805037.nc')
rad = myFile.metpy.parse_cf('Rad')
fk1 = float(myFile.metpy.parse_cf('planck_fk1'))
fk2 = float(myFile.metpy.parse_cf('planck_fk2'))
bc1 = float(myFile.metpy.parse_cf('planck_bc1'))
bc2 = float(myFile.metpy.parse_cf('planck_bc2'))
temp07 = (fk2/(xr.ufuncs.log(fk1/rad + 1)) - bc1)/bc2
np.save('../GOES_Files/test_files/OR_ABI-L1b-RadC-M3C07_G17_s20183171802219_e20183171805003_c20183171805037', create_npy(temp07, myFile))

myFile = xr.open_dataset('../GOES_Files/test_files/OR_ABI-L1b-RadC-M3C14_G17_s20183171802219_e20183171804592_c20183171805038.nc')
rad = myFile.metpy.parse_cf('Rad')
fk1 = float(myFile.metpy.parse_cf('planck_fk1'))
fk2 = float(myFile.metpy.parse_cf('planck_fk2'))
bc1 = float(myFile.metpy.parse_cf('planck_bc1'))
bc2 = float(myFile.metpy.parse_cf('planck_bc2'))
temp14 = (fk2/(xr.ufuncs.log(fk1/rad + 1)) - bc1)/bc2
np.save('../GOES_Files/test_files/OR_ABI-L1b-RadC-M3C14_G17_s20183171802219_e20183171804592_c20183171805038', create_npy(temp14, myFile))
```

```python
diff = temp07 - temp14

geos = diff.metpy.cartopy_crs
pc = ccrs.PlateCarree()

fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(projection=pc)

x = diff.x
y = diff.y

ax.set_extent([-124.50, -114.25, 32.50, 42.00], crs=pc)

ax.imshow(diff, origin='upper',
           extent=(x.min(), x.max(), y.min(), y.max()),
           transform=geos,
           interpolation='none',
           cmap = 'hot')
ax.add_feature(ccrs.cartopy.feature.STATES)
```
